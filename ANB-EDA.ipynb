**Task 4 - Data cleaning**

```python
data.isnull().sum()
# dropping all missing value rows
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.shape
```

```python
# Dealing with duplicates
data.duplicated().sum()

# deleting all duplicates rows
#data[data.duplicated()]
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# type casting
# changing data types

data['id'] = data['id'].astype(object)
data.dtypes
data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

# EDA

**Task 5 - Data Analysis**

**Univariate Analysis** - In EDA, univariate analysis examines a single column's distribution and key statistics.

```python
# identifying Outliers in price

#sns.boxplot(data=data,x='price')
df = data[data['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
# Price distribution
plt.figure(figsize=(9,5))
sns.histplot(data=df,x='price')
plt.ylabel('Frequency')
plt.xlabel('Price in $')
plt.title('Price Distribution')
plt.show()
```

```python
# Room Availability
plt.figure(figsize=(6,4))
sns.histplot(data=df,x='availability_365')
plt.ylabel('Frequency')
plt.xlabel('availability_365')
plt.title('availability_365 Distribution')
plt.show()
```

```python
df.dtypes
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
# price per beds
df['price_per_bed'] = df['price']/df['beds']
df.head()
```

```python
# Avg price per bad by neighbourhood_group
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

**BI Variable Analysis** - One variable depennency on another variable

```python
df.columns
```

```python
# price dependency on neighbourhood
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
# number o reviws and price realtion
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title("locality and review dependency")
plt.show()
```

```python
sns.pairplot(data=df,vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'],hue='room_type')
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x='longitude', y='latitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

```python
df.dtypes
```

```python
# heat map - correlation of one variable with others for numerical column

Numerical_col = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
Numerical_col
```

```python
sns.heatmap(data=Numerical_col,annot=True)
```
